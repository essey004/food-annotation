```python
import torch
torch.cuda.is_available()
from diffusers import AutoPipelineForText2Image
from PIL import Image
import matplotlib.pyplot as plt
import os

pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sd-turbo", torch_dtype=torch.float16)
pipe.to("cuda")

base = '/content/drive/MyDrive/food-annotation-project'
prompt_path = '/content/drive/MyDrive/food-annotation-project/prompts/prompts.txt'
output_path = '/content/drive/MyDrive/food-annotation-project/generated-data/set1'


with open(prompt_path, 'r') as f:
    prompts = [line.strip() for line in f if line.strip()]

for idx, prompt in enumerate(prompts, 1):
    print(f"Generating image {idx}/{len(prompts)}: {prompt}")
    image = pipe(prompt=prompt, num_inference_steps=4, guidance_scale=0.0).images[0]

    filename = f"food_{idx:03}.png"
    image.save(os.path.join(output_path, filename))
```
